#調整圖片大小統一為150*150

```python
def reshape_img(input_dir,output_dir):
  os.makedirs(output_dir, exist_ok=True)
  for filename in os.listdir(input_dir):
    # 拼接完整的文件路徑
    filepath = os.path.join(input_dir, filename)

    # 加載圖像
    image = Image.open(filepath)

    # 調整大小為目標尺寸
    target_size = (150, 150)
    resized_image = image.resize(target_size)

    # 調整大小後的圖像儲存到目標資料夾
    output_filepath = os.path.join(output_dir, filename)
    resized_image.save(output_filepath)
```

```python
input_dir = '/content/drive/My Drive/bravolav'
output_dir = '/content/drive/My Drive/CNN/train/Vivian'
reshape_img(input_dir,output_dir)
```

#資料擴增

將檔案中的圖片統一水平翻轉，擴展訓練集樣本

```python
import cv2
def data_additional(folder_path):
  image_files = os.listdir(folder_path)
  # 確定資料夾中只包含圖片檔案
  image_files = [file for file in image_files if file.endswith('.jpg')]
  n = 0
  # 處理每一張圖片，進行水平翻轉並保存到新的位置
  for file_name in image_files:
    # 讀取圖片
    n += 1
    image_path = os.path.join(folder_path, file_name)
    image = cv2.imread(image_path)

    # 水平翻轉圖片
    flipped_image = cv2.flip(image, 1)

    # 生成新的檔案名稱
    new_file_name = "flipped_" + file_name

    # 指定新檔案的保存路徑
    save_path = os.path.join(folder_path, new_file_name)

    # 將處理後的圖片保存到新的位置
    cv2.imwrite(save_path, flipped_image)
    print(f"水平翻轉並保存第{n}張圖片。")
```

```python
folder_path = '/content/drive/My Drive/CNN/train/Vivian'
data_additional(folder_path)
```

```python
folder_path = '/content/drive/My Drive/CNN/train/Kimi'
data_additional(folder_path)
```
